```
import requests
from bs4 import BeautifulSoup
import re
import datetime
import calendar

now = datetime.datetime.now()


def get_html(url):
    _html = ""
    resp = requests.get(url)
    if resp.status_code == 200:
        _html = resp.text
    return _html


def get_meal(code, ymd, weekday):
    schMmealScCode = code  # int=
    schYmd = ymd  # str

    num = weekday
    if weekday == "월":
        num = 1
    elif weekday == "화":
        num = 2
    elif weekday == "수":
        num = 3
    elif weekday == "목":
        num = 4
    elif weekday == "금":
        num = 5
    elif weekday == "토":
        num = 6
    elif weekday == "일":
        num = 7
    else:
        num = weekday

    URL = ("http://stu.sen.go.kr/sts_sci_md01_001.do?"
           "schulCode=B100000386&schulCrseScCode=4&schulKndScCode=04"
           "&schMmealScCode=%d&schYmd=%s" % (schMmealScCode, schYmd))
    html = get_html(URL)
    soup = BeautifulSoup(html, 'html.parser')
    element = soup.find_all("tr")
    element = element[2].find_all('td')
    try:
        element = element[num]  # num
        element = str(element)
        # element = element.replace('[', '')
        # element = element.replace(']', '')
        element = element.replace('<br/>', ' ')
        element = element.replace('<td class="textC last">', '')
        element = element.replace('<td class="textC">', '')
        element = element.replace('</td>', '')
        # element = element.replace('(h)', '')
        element = element.replace('.', '')
        element = element.replace('·', '')
        element = re.sub(r"\d", "", element)
    except:
        element = " "

    list_element = element.split(" ")
    list_element = list(filter(None, list_element))
    json_element = {}
    json_num = 0
    for i in list_element:
        json_element[json_num] = i
        json_num += 1

    return json_element

nowDate = now.strftime("%Y.%m.%d")  # 현재 y.m.d 불러오는 변수

weekday_dict = {0:'월', 1:'화',2:'수',3:'목',4:'금',5:'토',6:'일'}


sus=input('yyyy:')
'''
dnjf1=[]
def dnjf_wjdgkrl():
    dnjf=input('mm:')
    if dnjf == '01'or'02'or'03'or'04'or'05'or'06'or'07'or'08'or'09'or'10'or'11'or'12':
        dnjf1.append(0,dnjf)
    elif dnjf=='1':
        dnjf1.append(0,'01')
    elif dnjf=='2':
        dnjf1.insert(0,'02')
    elif dnjf=='3':
        dnjf1='03'
    elif dnjf=='4':
        dnjf1='04'
    elif dnjf=='5':
        dnjf1='05'
    elif dnjf=='6':
        dnjf1='06'
    elif dnjf=='7':
        dnjf1='07'
    elif dnjf=='8':
        dnjf1='08'
    elif dnjf=='9':
        dnjf1='09'
    else:
        print('잘못된 입력')
        dnjf_wjdgkrl()

dnjf_wjdgkrl()
'''
dnjf=input('mm:')
dlf=input('dd:')

dja=sus+'.'+dnjf+'.'+dlf

dydlf=weekday_dict[calendar.weekday(int(sus),int(dnjf),int(dlf))]
if dydlf =='토':
    print('토요일입니다')
elif dydlf =='일':
    print('일요일입니다')
else:
    meal=get_meal(2,dja,dydlf)
    print(meal)
```
